27. `17800`
> 首先使用类似杨辉三角形的方法可以知道 $P$ 到 $Q$ 的不同路线有 $200$ 条。
> 
> 本题的巧妙之处在于对每条线路，由于选择一次走一步还是两步带来的差异，按照路径中一次走两步的步骤（下称“大步”）的数量来罗列各种可能性，可以得到一般性结论如下：
> 
> 对长度为 $n$（简单起见假定 $n$ 为偶数，奇数的情况类似）的路径来说，有 $0$ 个大步的组合有 ${n \choose 0}$ 种，有 $1$ 个大步的组合有 ${{n-1} \choose 1}$ 种，有 $2$ 个大步的组合有 ${{n-2} \choose 2}$ 种，有 $3$ 个大步的组合有 ${{n-3} \choose 3}$ 种，依此类推，最后是有 $\frac{n}{2}$ 个大步的组合有 ${{\frac{n}{2}} \choose {\frac{n}{2}}}$ 种。
> 
> 按照本题所给图形，易得 $P$ $Q$ 之间距离长度为 $10$，故 $P$ $Q$ 之间的每条路线都有 ${10 \choose 0}+{9 \choose 1}+{8 \choose 2}+{7 \choose 3}+{6 \choose 4}+{5 \choose 5}=89$ 种走法。

30. $\frac{1960}{3}$
> 首先整理和分析题目描述的场景，可以注意到几个关键规律：1. ABCD 四个球在轨道上的位置顺序是不会变的；2. 任何时刻轨道上都有两个静止的球（分别位于 B 球和 D 球的初始位置，以下称“静止点”），一个顺时针运动的球，和一个逆时针运动的球；3. 顺时针运动的球速度一定是 1cm/s（以下称“慢球”），而逆时针运动的球速度一定是 29cm/s（以下称“快球”）。注意慢球和快球并不是固定的某个球，而是会随着碰撞换球，但任何时刻总有一个快球一个慢球分别在做固定速度固定方向的运动。
> 
> 接下来我们可以发现，“第一次三个小球相遇”必然发生在 B 球初始位置（以下称“关键相遇点”），静止在该位置上的球和两个正在运动的球在此相遇，此时慢球正好走过左下的四分之一圆弧，而快球已经转了好几圈了。进一步的分析可以发现，快球和慢球在左下四分之一圆弧内会有7次相遇，第8次相遇就会在关键相遇点发生，这7次相遇的位置与 A 球初始位置（以下称“原点”）的圆弧距离分别为 $\frac{20}{3}, \frac{60}{3}, \frac{100}{3}, \frac{140}{3}, \frac{180}{3}, \frac{220}{3}, \frac{260}{3}$。
> 
> 下一步需要画出这8次相遇时各球所在位置，在两个静止点的是哪两个球，在左下发生相遇的又是哪两个球，进而就可以计算 D 球在这过程中分别走了多远，最后算式为 $200+\frac{200}{3}+\frac{40}{3}+\frac{440}{3}+200+\frac{40}{3}+\frac{40}{3}$
> 
> 本题应该是全卷最有趣的一题。
